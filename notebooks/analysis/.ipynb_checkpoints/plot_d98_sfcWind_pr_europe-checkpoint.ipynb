```python
def pot(da,threshold,dim):
    
    assert (threshold>=0) & (threshold<1)
    
    return da.where(da>da.quantile(threshold,dim=dim))
```

```python
wind_dir = '/home/jovyan/CMIP6cf/output/timeseries/sfcWind_europe/'

source_ids = os.listdir(wind_dir) #intersection of models
ens_wind_ds={}
for s,source_id in enumerate(sorted([k for k in source_ids if ~k.startswith('.')])): #loop over models
    
    wind_ds = xr.open_mfdataset(os.path.join(wind_dir,source_id,'*ssp585*'))
    d98 = wind_ds.sfcWind.sel(time=slice('2071','2100')).quantile(.98,dim='time')-wind_ds.sfcWind.sel(time=slice('1991','2020')).quantile(.98,dim='time')
    if s==0:
        ens_wind_ds = d98
    else:
        ens_wind_ds = xr.concat((ens_wind_ds,d98),dim='source_id')
ens_wind_ds = ens_wind_ds.assign_coords({'source_id':sorted([k for k in source_ids if ~k.startswith('.')])})
    
'''   
wind = xr.open_dataset(os.path.join(wind_dir,'NorESM2-MM_gn_day_ssp585.nc')).sel(member_id='r1i1p1f1')
wind = wind.where(np.mod(wind.time.dt.month,12)<3)

wind_ens = xr.open_mfdataset(os.listdir('/home/jovyan/CMIP6cf/output/timeseries/sfcWind_europe')
'''
```

```python
pr_dir = '/home/jovyan/CMIP6cf/output/timeseries/pr_europe/'

source_ids = os.listdir(pr_dir) #intersection of models
ens_pr_ds={}
for s,source_id in enumerate(sorted([k for k in source_ids if ~k.startswith('.')])): #loop over models
    
    pr_ds = xr.open_mfdataset(os.path.join(pr_dir,source_id,'*ssp585*'))
    #pr_ds = pr_ds.where(np.mod(pr_ds.time.dt.month,12)<3)
    d98 = pr_ds.pr.sel(time=slice('2071','2100')).quantile(.98,dim='time')-pr_ds.pr.sel(time=slice('1991','2020')).quantile(.98,dim='time')
    if s==0:
        ens_pr_ds = d98
    else:
        ens_pr_ds = xr.concat((ens_pr_ds,d98),dim='source_id')
ens_pr_ds = ens_pr_ds.assign_coords({'source_id':sorted([k for k in source_ids if ~k.startswith('.')])})
    
```
