This notebook plots the dependence-driven changes in the number of joint wind speed & precipitation extremes and the changes in the univariate wind speed & precipitation extremes in winter and autumn (DJF & SON).

```python
import numpy as np
import xarray as xr
import os
import fnmatch
import matplotlib.pyplot as plt
import cartopy
import cartopy.crs as ccrs
from mpl_toolkits.axes_grid1.inset_locator import inset_axes
%config InlineBackend.figure_format = 'retina'
```

```python
def preprocess(ds):
    ds = ds.expand_dims(dim={"source_id": [ds.source_id]})
    return ds
```

```python
def subplot_n_wp_eu(da,gridspec_idx,subplot_title,cmap,v,plot_cbar,cbar_label=None):
    ax = plt.subplot(gs[gridspec_idx[0],gridspec_idx[-1]],projection=ccrs.Robinson(central_longitude=0))

    ax.add_feature(cartopy.feature.OCEAN, zorder=0,facecolor='lightgrey')
    ax.add_feature(cartopy.feature.LAND, zorder=0, facecolor='grey')
    
    sc=da.plot(cmap=cmap,vmin=v[0],vmax=v[-1],transform=ccrs.PlateCarree(),zorder=3,ax=ax,add_colorbar=False)
    sc.set_edgecolor('face')
    ax.coastlines(zorder=5)
    ax.set_extent([-24, 13, 35, 65], crs=ccrs.PlateCarree())
    if plot_cbar:
        cax=ax.inset_axes(bounds=(0, -.1,1,.075))
        cb=fig.colorbar(sc, cax=cax,orientation='horizontal',label=cbar_label)
    ax.set_title(subplot_title)
    return ax,sc
```

Load output of joint probability analysis:

```python
in_dir = '/home/jovyan/CMIP6cex/output/num_extremes/timeseries_eu_1p5/sfcWind_pr/40yr_p98_lag0d_declus1d_ref2000/'
ssps = ['ssp245','ssp585']

ds_fns = [] #get filenames wind & pr
for top, dirs, files in os.walk(in_dir):
    for file in files:
           ds_fns.append(os.path.join(top, file))
            
ds_surge_fns = [] #get filenames surge & pr
for top, dirs, files in os.walk('/home/jovyan/CMIP6cex/output/num_extremes/timeseries_eu_gesla2_tgs/surge_pr/40yr_p98_lag0d_declus1d_ref2000/'):
    for file in files:
           ds_surge_fns.append(os.path.join(top, file))          

#we have more availability for sfcWind than surge (also needs psl), only open the overlapping files here
ssps_ds = []
for ssp in ssps:
    ssp_fns = fnmatch.filter(ds_fns,'*'+ssp+'*')
    ssp_surge_fns = fnmatch.filter(ds_surge_fns,'*'+ssp+'*')
    ssp_fns = [f for f in ssp_fns if '/'.join(f.split('/')[-3::]) in ['/'.join(k.split('/')[-3::]) for k in ssp_surge_fns]]
    ssp_ds = xr.open_mfdataset((ssp_fns),preprocess=preprocess,coords='minimal').chunk({'member_id':200})
    ssps_ds.append(ssp_ds)

numex_ds = xr.concat(ssps_ds,dim='ssp')
numex_ds = numex_ds.assign_coords({'ssp':ssps})

#load data to compute dependence-driven changes & changes in number of univariate extremes:
num_joint_wp = numex_ds[['num_joint_extremes','num_pr_extremes',
                              'num_sfcWind_extremes','num_joint_extremes_sfcWind_pr_driven','num_sfcWind_extremes_refWindow_futT','num_pr_extremes_refWindow_futT']].load()
num_joint_wp_isect = num_joint_wp.where(np.isfinite(num_joint_wp.num_joint_extremes).all(dim=['ssp','latitude','longitude']),drop=True) #members with both SSPs available
```

```python
fig=plt.figure(figsize=(8,5)) #generate figure  
gs = fig.add_gridspec(2,3)
gs.update(hspace=.1)

ssp='ssp585'

for season in np.arange(2):
    #ssp_ds = ds_intersect.sel(ssp=ssp).sum(dim='month',skipna=False)
    if season ==0: #djf
        numex = num_joint_wp_isect.sel(ssp=ssp).isel(month=-1) + num_joint_wp_isect.sel(ssp=ssp).isel(month=0) + num_joint_wp_isect.sel(ssp=ssp).isel(month=1)
    elif season==1:
        numex = num_joint_wp_isect.sel(ssp=ssp).isel(month=8) + num_joint_wp_isect.sel(ssp=ssp).isel(month=9) + num_joint_wp_isect.sel(ssp=ssp).isel(month=10) 

    dnumex = numex.sel(window=2080) - numex.sel(window=2000)
    dependence_driven = dnumex.num_joint_extremes - dnumex.num_joint_extremes_sfcWind_pr_driven
    if season ==0:
        ax,sc = subplot_n_wp_eu(.25*dependence_driven.mean(dim='member_id').mean(dim='source_id'),[season,0],
                        ['(a)','(d)'][season],'seismic',[-10,10],plot_cbar=False)
        ax.annotate('DJF',xy=(-.1,.46),xycoords='axes fraction',rotation=90)
        
        ax,sc = subplot_n_wp_eu(.25*(numex.num_pr_extremes.sel(window=2080)-numex.num_pr_extremes_refWindow_futT.sel(window=2080)).mean(dim='member_id').mean(dim='source_id'),[season,2],
                        ['(c)','(f)'][season],'seismic',[-10,10],plot_cbar=False)
        
        ax,sc = subplot_n_wp_eu(.25*(numex.num_sfcWind_extremes.sel(window=2080)-numex.num_sfcWind_extremes_refWindow_futT.sel(window=2080)).mean(dim='member_id').mean(dim='source_id'),[season,1],
                        ['(b)','(e)'][season],'seismic',[-10,10],plot_cbar=False)
        
        
    else:
        ax,sc = subplot_n_wp_eu(.25*dependence_driven.mean(dim='member_id').mean(dim='source_id'),[season,0],
                        ['(a)','(d)'][season],'seismic',[-10,10],plot_cbar=True,cbar_label='$\Delta N_{W \wedge P}$ [#/decade]')
        ax.annotate('SON',xy=(-.1,.43),xycoords='axes fraction',rotation=90)
        
        ax,sc = subplot_n_wp_eu(.25*(numex.num_pr_extremes.sel(window=2080)-numex.num_pr_extremes_refWindow_futT.sel(window=2080)).mean(dim='member_id').mean(dim='source_id'),[season,2],
                        ['(c)','(f)'][season],'seismic',[-10,10],plot_cbar=True,cbar_label='$\Delta N_{P}$ [#/decade]')
        
        ax,sc = subplot_n_wp_eu(.25*(numex.num_sfcWind_extremes.sel(window=2080)-numex.num_sfcWind_extremes_refWindow_futT.sel(window=2080)).mean(dim='member_id').mean(dim='source_id'),[season,1],
                        ['(b)','(e)'][season],'seismic',[-10,10],plot_cbar=True,cbar_label='$\Delta N_{W}$ [#/decade]')
        
#fig.savefig('Figures/seasonal_dependence_v2.pdf',format='pdf',dpi=300,bbox_inches='tight')
```
